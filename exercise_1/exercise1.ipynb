Import Libraries

```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
import geopandas as gpd
```

Import data

```python
df = sm.datasets.get_rdataset('EuroEnergy', 'AER', cache=True).data
df.reset_index(inplace=True)
display(df)
```

```python
"https://www.naturalearthdata.com/downloads/110m-cultural-vectors/"
#"C:\Users\maxnd\Documents\Machine Learning\pythonfordataanalysis\exercise_1\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp"
#"exercise_1\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp"

world = gpd.read_file(r"exercise_1\Natural Earth 110m Countries\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp")
world = world[world["CONTINENT"]=="Europe"]
world.columns = world.columns.str.lower()

display(world)
```

```python
world.columns
```

Data Cleaning

```python
df.rename(columns={"rownames":"country"}, inplace=True)
```

```python
# Renaming WGermany to Germany & UK to United Kingdom
df["country"].replace({"WGermany":"Germany", "UK":"United Kingdom"}, inplace=True)
```

```python
bins = [0, 5000, 30000, df["energy"].max()]
labels = ["lower", "medium", "high"]

display(df)
```

```python
df["energy_category"] = pd.cut(df["energy"], bins=bins, labels=labels)
display(df)
```

Combining data source

```python
df = world.merge(df, left_on="name", right_on="country", how="left")
df = df[["name", "energy", "economy", "geometry", "energy_category"]]
```

Filling null values in the energy column

why do we need to fill null values

-> ensuring consistency and accuracy in our analysis

```python
# Checking how many null values we have in the energy coloumn
df["energy"].isna().sum()

# Filling the null values
#df["energy"].fillna(0, inplace=True)
```

```python
# Filling the null values
df["energy"].fillna(0, inplace=True)
```

```python
df["energy"].isna().sum() 
```

result indicating we have no null values in the coloumn "energy"
