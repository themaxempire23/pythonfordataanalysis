Analysing the data using correlation, I've seen loads of null values

```python
def extract_data(sheet, country, column_name):
    df = pd.read_excel(file, sheet_name=sheet, skiprows=2)
    df.rename(columns={df.columns[0]: "country"}, inplace=True)
    df.set_index("country", inplace=True)
    df.replace({"-": 0}, inplace=True)

    df = pd.DataFrame(df.loc[country])
    df.rename(columns={country: column_name}, inplace=True)

    return df
```

```python
carbon = extract_data("CO2e Emissions ", "Total World", "carbon")
Solar = extract_data('Solar Generation - TWh', "Total World", "Solar")
Wind = extract_data('Wind Generation - TWh', "Total World", "Wind")
Nuclear = extract_data("Nuclear Generation - TWh", "Total World", "Nuclear")
Biomass = extract_data("Geo Biomass Other - TWh", "Total World", "Biomass")
Hydro = extract_data('Hydro Generation - TWh', "Total World", "Hydro")
```

```python
data = pd.concat([carbon, Solar, Wind, Nuclear, Biomass, Hydro], axis=1)
```

```python
sns.heatmap(data.corr(), cmap="crest")
```
