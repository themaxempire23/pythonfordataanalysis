Dataset


```python
data = {'Name': ["Simson", "Anna", "Peter", "Linda", "Maria", "Daniel", "Simon", "Monica"],
        'Profession':["Data Analyst", "Steward", "Refree", "Mechanic", "Pilot", "Nurse", "Police officer", "Doctor"],
        'Age':[28, 25, 38, 32, 55, 45, 59, 34],
        'Country':["Namibia", "South Africa", "USA", "Canada", "Nigeria", "Germany", "Germany", "Portugal"]}

df = pd.DataFrame(data)
```

Saving dataset into csv file

```python
df.to_csv("Regression_dset.csv", index=False)
```

Using the country coloumn

```python
# creating a new coloumn
df["Country_encoder"] = df["Country"].astype("category").cat.codes
```

```python
pd.read_csv('Regression_dset.csv')
```

```python
X = df["Country_encoder"]
y = df["Age"]
pd.read_csv('Regression_dset.csv')
```

way of adding a new constant

```python
X = df["Country_encoder"]
y = df["Age"]

X = sm.add_constant(X)

#developing model
model = sm.OLS(y, X).fit()

model.summary()
```

Multiple regression analysis

```python
# working with multiple values

df["Name_encode"] = df["Name"].astype("category").cat.codes
pd.read_csv('Regression_dset.csv')
```

```python
X = df[["Country_encoder", "Name_encode"]]

X = sm.add_constant(X)

#developing model
model = sm.OLS(y, X).fit()

model.summary()
```
