Import libraries

```python
import seaborn as sns
import statsmodels.api as sm
import pandas as pd
```

Import dataset

```python
co2 = sm.datasets.co2.load_pandas()
co2 = co2.data
```

Checking for null values in co2 dataset

```python
co2.isna().sum().sum()
```

```python
# Checking the Structure of the data

# 1st reseting the indez
co2.reset_index(inplace=True)

# renaming index column
co2.rename(columns={"index":"date"}, inplace=True)
```

Setting up a time serie


```python
co2["month/year"]=co2["date"].dt.strftime("%m/%y")
co2["month"] = co2["date"].dt.strftime("%m")
```

Filling the data, (filling the data using interpolate method)

```python
co2["new_data"] = co2["co2"].interpolate(method="polynomial", order=3, limit_direction="both")
```

Visualing the data

```python
# Lineplot
sns.lineplot(data=co2, x="date", y="co2", hue="month")
```

```python
# barplot , to see mean values
sns.barplot(data=co2, x="month", y="co2", estimator="mean")
```
