```python
import pandas as pd
import numpy as np
```

Time series DataFrame

```python
df = pd.DataFrame({
    'Date': pd.date_range(start='2025-07-16', periods=60),
    'Value': np.random.randint(0, 100, size=60)
})
```

```python
display(df)
```

```python
# Exracting Month/Year base value

df["Month/Year"] = df["Date"].dt.strftime("%b/%Y")
display(df)
```

```python
# Extracting weekly values
df["Week/Year"] = df["Date"].dt.strftime("%U/%Y")
display(df)
```

```python
# Getting mean values per week
df.groupby("Week/Year") ["Value"].mean()
```

```python
# Getting max values per week
df.groupby("Week/Year") ["Value"].max()
```

```python
#Getting the mean, sum and stdv in the dataset
df.groupby("Week/Year")["Value"].agg(["mean", "sum", "std"])
```
