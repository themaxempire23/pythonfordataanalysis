```python
import seaborn as sns
```

```python
sns.get_dataset_names() # To check different datasets in seaborn
```

```python
sns.load_dataset("flights")
```

```python
flights = sns.load_dataset("flights")
```

line chart

```python
sns.lineplot(data=flights,
            x="year",
            y="passengers",
            errorbar=None,
            hue="month",
            #hue="year"
            )
```

Barchart

```python
# 1. Vertical Barchart

sns.barplot(data=flights,
            x="month",
            y="passengers",
            errorbar=None,
            estimator="sum")
```

```python
# 2. Horizontal Barchart

sns.barplot(data=flights,
            y="month",
            x="passengers",
            errorbar=None,
            estimator="sum", orient="y")

            # Result: Transpose the visualizationo of the data
```

Scatter plot: Heavily used when we want to compare different information

```python
sns.scatterplot(data=flights,
                x="year",
                y="passengers")
```

 Heatmaps

 Mostly use in corrilation of data

```python
sns.heatmap(flights.pivot(index="year", columns="month", values="passengers"),
cmap="crest")
```

Histogram

```python
sns.histplot(data=flights,
             x="passengers")
```

```python
sns.histplot(data=flights,
             x="passengers",
             bins=30,
             kde=True) #kde= smootline related to bar
```

```python
sns.histplot(data=flights,
             x="passengers",
             bins=30,
             hue="month",
             multiple="stack")
```

Boxplot/Violin

```python
sns.boxplot(data=flights,
             y="passengers"
             #x="year"
             )
```

```python
sns.boxplot(data=flights,
             y="passengers",
             x="year"
             )
```

```python
#Violin

sns.violinplot(data=flights,
                y="passengers")
```

```python

sns.violinplot(data=flights,
                y="passengers",
                x="year")
```
